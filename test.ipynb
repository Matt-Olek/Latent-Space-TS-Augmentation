```python
from sklearn.mixture import GaussianMixture
import numpy as np


random_data = np.random.rand(100,2)
random_data2 = np.random.rand(100,2)
random_data3 = np.random.rand(100,2)

gmm1 = GaussianMixture(n_components=5)
gmm1.fit(random_data)

gmm2 = GaussianMixture(n_components=7)
gmm2.fit(random_data)

gmm3 = GaussianMixture(n_components=3)
gmm3.fit(random_data)
```

```python
# Sample from the first GMM

sampled_data, _ = gmm1.sample(10)
```

```python
score_1 = gmm1.score_samples(sampled_data)
score_2 = gmm2.score_samples(sampled_data)
score_3 = gmm3.score_samples(sampled_data)
```

```python
import torch

log_probs = torch.stack([torch.tensor(score_1), torch.tensor(score_2), torch.tensor(score_3)], dim=1)
```

```python
proportions = (log_probs[:, 0].unsqueeze(1) < log_probs).float().mean(dim=0)
```
